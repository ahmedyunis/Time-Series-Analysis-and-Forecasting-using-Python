# load Dataset 

```python
df = pd.read_csv('D:/TimeSeries/Time Series Analysis and Forcasting usin Pyhton/Forecasting Resources (ST Academy)\Data + Code/us-airlines-monthly-aircraft-miles-flown.csv',header=0,parse_dates=[0])
```

# Downsampling 

```python
Quart_df = df.resample('Q', on='Month').mean()
```

```python
Quart_df.head()
```

```python
year_df = df.resample('A', on='Month').sum()
```

```python
year_df.head()
```

# UpSampling 

```python
upsample_df = df.resample('D',on='Month').mean()
```

```python
upsample_df.head(50)
```

```python
interplate_df=upsample_df.interpolate(method='linear')
```

```python
interplate_df.head(50)
```

```python
interplate_df.plot()
```

```python
poly_df = upsample_df.interpolate(method='spline', order=2)
```

```python
poly_df.head()
```

```python
poly_df.plot()
```
