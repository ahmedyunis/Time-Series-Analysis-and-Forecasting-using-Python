# Import Libraries 

```python
import pandas as pd
```

# Load DataSet 

```python
datafram2 = pd.read_csv('D:/TimeSeries/Time Series Analysis and Forcasting usin Pyhton/Forecasting Resources (ST Academy)\Data + Code/daily-total-female-births-CA.csv',
                       header=0,parse_dates=[0])
```

```python
datafram2.head()
```

```python
datafram2.date.dtypes
```

# Date time Features

```python
df_features = datafram2.copy()
```

```python
df_features['year'] = datafram2.date.dt.year
```

```python
df_features['month'] = datafram2.date.dt.month
```

```python
df_features['day'] = datafram2.date.dt.day
```

```python
df_features.head()
```

# Lag Features

```python
df_features['lag1']=datafram2.births.shift(1)
```

```python
df_features['lag2']=datafram2.births.shift(365)
```

```python
df_features.head()
```

# Windows Features

```python
df_features['Roll_mean'] = datafram2.births.rolling(window=2).mean()
```

```python
df_features.head()
```

```python
df_features['Roll_max'] = datafram2.births.rolling(window=3).max()
```

```python
df_features.head()
```

# Expanding Features

```python
df_features['Expand_max'] = datafram2.births.expanding().max()
```
